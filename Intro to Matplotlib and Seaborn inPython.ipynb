
**`STAKED BAR PLOT`**

---

```python
#sample data frame
data = pd.DataFrame([[1, 2],
                    [3, 4],
                    [8, 7],
                    [6, 5],
                    [4, 3],
                    [2, 1]])

data.plot(kind='bar', stacked=True);
```

**`Visualization With Seaborn`**

---
---

- Seaborn is a Pyhton Data visualization linbrary based on Matplotlib.

- It provides a high-level interface for drawing attractive and informative statistical graphics. It provides choice for plot style and color default, defines simple high-level functions for common statistical plot types, and integrated with the functinality provided by Pandas DataFrames.

- The main idea of Seaborn is that it provides high-level commands to create a vareity of plot types useful for statiscal exploration and even some statistical model fitting.

```python
import seaborn as sns
sns.set()
```

```python
# create a random data
data = np.random.multivariate_normal([0,0], [[5,2],[2,2]], size=2000)
data =  pd.DataFrame(data, columns=['x', 'y'])
```

```python
data
```

---
---

**`DENSITY PLOT`**

---

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

```python
for col in 'xy':
    sns.distplot(data[col], hist=False)
```

```python
for col in 'xy':
    sns.distplot(data[col])
```


**`PAIR PLOT`**

---

```python
iris = sns.load_dataset("iris")
iris.head()
```

***Visualizing the multidimensional relationships among the samples is as easy as calling sns.pairplot:***

```python
sns.pairplot(iris, hue='species', height=2.5);
```
